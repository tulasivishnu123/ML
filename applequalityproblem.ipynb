```python
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd

# Create a sample dataset
X = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])  # Feature data
y = np.array([0, 1, 0, 1])  # Labels

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Print the split sets
print("X_train: ", X_train)
print("X_test: ", X_test)
print("y_train: ", y_train)
print("y_test: ", y_test)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
import pandas as pd

# Create a sample dataset
text = ['Hello my name is james', 'james this is my python notebook', 'james trying to create a big project']

# Create a CountVectorizer object
vectorizer = CountVectorizer()

# Fit the vectorizer to the text data
vectorizer.fit(text)

# Transform the text data into a matrix of token counts
X = vectorizer.transform(text)

# Print the resulting matrix
print(X)
```

```python
import pandas as pd
data=pd.read_csv('applequality.csv')
data
```

```python
df=pd.DataFrame(data)
df=df.drop('A_id',axis=1)
df
df=df.drop(4000,axis=0)
df
feature=df[{'Size','Weight','Sweetness','Crunchiness','Juiciness','Ripeness','Acidity'}]
classlabel=df[{'Quality'}]
print(feature)
print(classlabel)
```

```python
X_train, X_test,Y_tr,Y_te = train_test_split(feature,classlabel,test_size=0.25)
X_train
```

```python
X_test
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train,Y_tr)
y=clf.predict(X_test)
accuracy = clf.score(X_test, Y_te)
print("Accuracy:", accuracy)
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix
confusionmatrix = confusion_matrix(Y_te, y)
print("Confusion matrix:\n%s" % confusionmatrix)
```
